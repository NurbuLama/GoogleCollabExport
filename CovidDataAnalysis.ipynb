```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
%matplotlib inline
```

```
df= pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv')
```

```
df.info()
```

```
df.head()
```

```
drop_this=['fips']
df.drop(drop_this, inplace= True, axis=1)
```

```
df.head()
```

```
df['date']= pd.to_datetime(df['date'], format='%Y-%m-%d')
```

```
df.info()
```

```
missing_values=['n/a', 'na', 'nan', '--']
```

```
df.fillna(value=-1, inplace=True)
```

```
df.groupby('state').count()['cases'].sort_values(ascending= True)
```

```
df1= df[df['state']=='New York']
```

```
df1.tail()
```

```
plt.figure(figsize=(10,6))
sns.set_style('whitegrid')
df1['deaths'].hist(bins=10)
plt.xlabel('deaths')
```

```
sns.jointplot(x='county', y='cases', data=df1)
```

```
sns.lineplot(x = 'date', y = 'cases', data = df1).set_title('New York State Cases')
plt.xticks(rotation=45, ha='right')
plt.show()

sns.lineplot(x = 'date', y = 'cases', data = df[df['state']=='California']).set_title('California State Cases')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```
sns.lineplot(data = df[df['state']=='California'], x = 'date', linewidth = 2.5, y = 'deaths',hue = 'state').set_title('State Wise Comparision ')
plt.xticks(rotation = 45, ha = 'right')
plt.show()

sns.lineplot(data = df[df['state']=='Texas'], x = 'date', linewidth = 2.5, y = 'deaths',hue = 'state').set_title('Sussex & Kent COVID Deaths')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```
